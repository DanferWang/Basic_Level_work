```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# GlobalModel Test
## kappa

```python
global_res = np.load('./resultDict/GlobalModel_Test_kappa.npy', allow_pickle=True)
global_res_kappa= global_res.item()

global_res_kappa
```

```python
time_list = ['1y', '5y', '10y', '20y', '50y', '100y', '200y', '400y', '500y']
```

```python
def plot_res(res_dict, label, title):
    fig, ax = plt.subplots()
    fig.set_facecolor('white')
    ax.plot(time_list, res_dict['mean'], label='mean', marker='.', color='#2391FF')
    ax.plot(time_list, res_dict['max'], label='max', marker='.', color='#EB4185')

    lim_l = min(res_dict['mean'] + res_dict['max']) - 0.005
    lim_h = max(res_dict['mean'] + res_dict['max']) + 0.005
    ax.set(xlabel='n resent years', ylabel=label, ylim=[lim_l, lim_h])

    ax.set_title('Ngram ' + title + ' ' + label)
    fig.legend()

    plt.savefig('./readme.assets/'+title[0:-6]+'_'+label+'.png')
    plt.show()
```

```python
plot_res(global_res_kappa, 'kappa', 'GlobalModel Test')
```

## balanced accuracy

```python
global_res_acc = np.load('./resultDict/GlobalModel_Test_balanced_acc.npy', allow_pickle=True)
global_res_acc = global_res_acc.item()
```

```python
plot_res(global_res_acc, 'balanced accuracy', 'GlobalModel Test')
```
