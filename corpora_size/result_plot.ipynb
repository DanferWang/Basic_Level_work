```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# kappa
x = ['1M', '2.4M', '5.7M', '100M']
BNC = [0.687879, 0.684052, 0.689859, 0.686256]
CHILDS = [0.665358, 0.669301, 0.673659, np.nan]
CABNC = [0.674959, 0.687207, np.nan, np.nan]
KBNC = [0.687594, np.nan, np.nan, np.nan]
NGrams = [0.677357, 0.677357, 0.677357, 0.677357]

fig, ax = plt.subplots()
fig.set_facecolor('white')
ax.plot(x, BNC, label='BNC', marker='.', color='#5B8FF9')
ax.plot(x, CHILDS, label='CHILDS', marker='.', color='#61DDAA')
ax.plot(x, CABNC, label='CABNC', marker='.', color='#F6BD16')
ax.plot(x, KBNC, label='KBNC', marker='.', color='#9661BC')
ax.plot(x, NGrams, label='NGrams', marker='.', color='#F08BB4')

ax.set(xlabel='Corpora size', ylabel='kappa', ylim=[0.65, 0.72])

fig.legend()
plt.show()
```
