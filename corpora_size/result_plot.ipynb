## 50 times samplings and GlobalModel test

```python
# KBNC
kbnc_res = np.load('./several_times_sampling/GlobalModelTest/50_times_sampling_KBNC.npy', allow_pickle=True).item()
res_df = pd.DataFrame.from_dict(kbnc_res)
kappa_df = pd.DataFrame()
balanced_acc_df = pd.DataFrame()
for index, row in res_df.iterrows():
    for counter in range(50):
        kappa_df.at[index, counter] = row[counter]['kappa']
        balanced_acc_df.at[index, counter] = row[counter]['balanced acc']

kbnc_1m_kappa = np.array(kappa_df[0:1]).transpose().mean()
kbnc_1m_acc = np.array(balanced_acc_df[0:1]).transpose().mean()

(kbnc_1m_kappa, kbnc_1m_acc)
```

```python
# CABNC
cabnc_res = np.load('./several_times_sampling/GlobalModelTest/50_times_sampling_CABNC.npy', allow_pickle=True).item()
res_df = pd.DataFrame.from_dict(cabnc_res)
kappa_df = pd.DataFrame()
balanced_acc_df = pd.DataFrame()
for index, row in res_df.iterrows():
    for counter in range(50):
        kappa_df.at[index, counter] = row[counter]['kappa']
        balanced_acc_df.at[index, counter] = row[counter]['balanced acc']

cabnc_1m_kappa = np.array(kappa_df[0:1]).transpose().mean()
cabnc_2_4m_kappa = np.array(kappa_df[1:2]).transpose().mean()
cabnc_1m_acc = np.array(balanced_acc_df[0:1]).transpose().mean()
cabnc_2_4m_acc = np.array(balanced_acc_df[1:2]).transpose().mean()

(cabnc_1m_kappa, cabnc_2_4m_kappa, cabnc_1m_acc, cabnc_2_4m_acc)
```

```python
# CHILDES
childes_res = np.load('./several_times_sampling/GlobalModelTest/50_times_sampling_CHILDES.npy', allow_pickle=True).item()
res_df = pd.DataFrame.from_dict(childes_res)
kappa_df = pd.DataFrame()
balanced_acc_df = pd.DataFrame()
for index, row in res_df.iterrows():
    for counter in range(50):
        kappa_df.at[index, counter] = row[counter]['kappa']
        balanced_acc_df.at[index, counter] = row[counter]['balanced acc']

childes_1m_kappa = np.array(kappa_df[0:1]).transpose().mean()
childes_2_4m_kappa = np.array(kappa_df[1:2]).transpose().mean()
childes_5_7m_kappa = np.array(kappa_df[2:3]).transpose().mean()
childes_1m_acc = np.array(balanced_acc_df[0:1]).transpose().mean()
childes_2_4m_acc = np.array(balanced_acc_df[1:2]).transpose().mean()
childes_5_7m_acc = np.array(balanced_acc_df[2:3]).transpose().mean()

(childes_1m_kappa, childes_2_4m_kappa, childes_5_7m_kappa, childes_1m_acc, childes_2_4m_acc, childes_5_7m_acc)
```

```python
# BNC
bnc_res = np.load('./several_times_sampling/GlobalModelTest/50_times_sampling_BNC.npy', allow_pickle=True).item()
res_df = pd.DataFrame.from_dict(bnc_res)
kappa_df = pd.DataFrame()
balanced_acc_df = pd.DataFrame()
for index, row in res_df.iterrows():
    for counter in range(50):
        kappa_df.at[index, counter] = row[counter]['kappa']
        balanced_acc_df.at[index, counter] = row[counter]['balanced acc']

bnc_1m_kappa = np.array(kappa_df[0:1]).transpose().mean()
bnc_2_4m_kappa = np.array(kappa_df[1:2]).transpose().mean()
bnc_5_7m_kappa = np.array(kappa_df[2:3]).transpose().mean()
bnc_100m_kappa = np.array(kappa_df[3:4]).transpose().mean()
bnc_1m_acc = np.array(balanced_acc_df[0:1]).transpose().mean()
bnc_2_4m_acc = np.array(balanced_acc_df[1:2]).transpose().mean()
bnc_5_7m_acc = np.array(balanced_acc_df[2:3]).transpose().mean()
bnc_100m_acc = np.array(balanced_acc_df[3:4]).transpose().mean()

(bnc_1m_kappa, bnc_2_4m_kappa, bnc_5_7m_kappa, bnc_100m_kappa, bnc_1m_acc, bnc_2_4m_acc, bnc_5_7m_acc, bnc_100m_acc)
```
