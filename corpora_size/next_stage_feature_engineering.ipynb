```python
import pandas as pd
```

## max, min and sum occurrences in the same size among corpora

```python
first_stage = pd.read_csv('./all_occurrence_size_differential_features.csv')

first_stage
```

```python
size_group_1m = ['bnc_1m_sum', 'childes_1m_sum', 'cabnc_1m_sum', 'kbnc_1m_sum']
size_group_2_4m = ['bnc_2_4m_sum', 'childes_2_4m_sum', 'cabnc_2_4m_sum']
size_group_5_7m = ['bnc_5_7m_sum', 'childes_5_7m_sum']
size_group_100m = ['bnc_100m_sum']

next_stage = first_stage

# max
next_stage['max_1m'] = next_stage[size_group_1m].max(axis=1)
next_stage['max_2_4m'] = next_stage[size_group_2_4m].max(axis=1)
next_stage['max_5_7m'] = next_stage[size_group_5_7m].max(axis=1)
next_stage['max_100m'] = next_stage[size_group_100m].max(axis=1)

# min
next_stage['min_1m'] = next_stage[size_group_1m].min(axis=1)
next_stage['min_2_4m'] = next_stage[size_group_2_4m].min(axis=1)
next_stage['min_5_7m'] = next_stage[size_group_5_7m].min(axis=1)
next_stage['min_100m'] = next_stage[size_group_100m].min(axis=1)

# sum
next_stage['sum_1m'] = next_stage[size_group_1m].sum(axis=1)
next_stage['sum_2_4m'] = next_stage[size_group_2_4m].sum(axis=1)
next_stage['sum_5_7m'] = next_stage[size_group_5_7m].sum(axis=1)
next_stage['sum_100m'] = next_stage[size_group_100m].sum(axis=1)

next_stage
```

## avg occurrences in each size

```python
corpus_bnc = ['bnc_1m_sum', 'bnc_2_4m_sum', 'bnc_5_7m_sum', 'bnc_100m_sum']
corpus_childes = ['childes_1m_sum', 'childes_2_4m_sum', 'childes_5_7m_sum']
corpus_cabnc = ['cabnc_1m_sum', 'cabnc_2_4m_sum']
corpus_kbnc = ['kbnc_1m_sum']

# avg
next_stage['avg_bnc'] = next_stage[corpus_bnc].mean(axis=1)
next_stage['avg_childes'] = next_stage[corpus_childes].mean(axis=1)
next_stage['avg_cabnc'] = next_stage[corpus_cabnc].mean(axis=1)
next_stage['avg_kbnc'] = next_stage[corpus_kbnc].mean(axis=1)

next_stage
```

```python
next_stage.to_csv('./next_stage_size_differential_features.csv', index=False)
```
