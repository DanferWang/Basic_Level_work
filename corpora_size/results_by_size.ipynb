```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## GlobalModel

```python
# load sampling results
kbnc_res = np.load('./several_times_sampling/GlobalModelTest/50_times_sampling_KBNC.npy', allow_pickle=True).item()
cabnc_res = np.load('./several_times_sampling/GlobalModelTest/50_times_sampling_CABNC.npy', allow_pickle=True).item()
childes_res = np.load('./several_times_sampling/GlobalModelTest/50_times_sampling_CHILDES.npy', allow_pickle=True).item()
bnc_res = np.load('./several_times_sampling/GlobalModelTest/50_times_sampling_BNC.npy', allow_pickle=True).item()
```

```python
kbnc_df = pd.DataFrame.from_dict(kbnc_res)
kbnc_kappa_df = pd.DataFrame()
kbnc_balanced_acc_df = pd.DataFrame()
for index, row in kbnc_df.iterrows():
    for counter in range(50):
        kbnc_kappa_df.at[index, counter] = row[counter]['kappa']
        kbnc_balanced_acc_df.at[index, counter] = row[counter]['balanced acc']

cabnc_df = pd.DataFrame.from_dict(cabnc_res)
cabnc_kappa_df = pd.DataFrame()
cabnc_balanced_acc_df = pd.DataFrame()
for index, row in cabnc_df.iterrows():
    for counter in range(50):
        cabnc_kappa_df.at[index, counter] = row[counter]['kappa']
        cabnc_balanced_acc_df.at[index, counter] = row[counter]['balanced acc']

childes_df = pd.DataFrame.from_dict(childes_res)
childes_kappa_df = pd.DataFrame()
childes_balanced_acc_df = pd.DataFrame()
for index, row in childes_df.iterrows():
    for counter in range(50):
        childes_kappa_df.at[index, counter] = row[counter]['kappa']
        childes_balanced_acc_df.at[index, counter] = row[counter]['balanced acc']

bnc_df = pd.DataFrame.from_dict(bnc_res)
bnc_kappa_df = pd.DataFrame()
bnc_balanced_acc_df = pd.DataFrame()
for index, row in bnc_df.iterrows():
    for counter in range(50):
        bnc_kappa_df.at[index, counter] = row[counter]['kappa']
        bnc_balanced_acc_df.at[index, counter] = row[counter]['balanced acc']
```

### 1M

```python
# kappa
kbnc_kappa_1m = np.array(kbnc_kappa_df[0:1]).transpose()
cabnc_kappa_1m = np.array(cabnc_kappa_df[0:1]).transpose()
childes_kappa_1m = np.array(childes_kappa_df[0:1]).transpose()
bnc_kappa_1m = np.array(bnc_kappa_df[0:1]).transpose()

fig, ax = plt.subplots()
fig.set_facecolor('white')
fig.set_figwidth(12)
x = np.array([x+1 for x in range(50)])
ax.plot(x, bnc_kappa_1m, label='BNC', marker='.', color='#5B8FF9')
ax.plot(x, childes_kappa_1m, label='CHILDES', marker='.', color='#61DDAA')
ax.plot(x, cabnc_kappa_1m, label='CABNC', marker='.', color='#F6BD16')
ax.plot(x, kbnc_kappa_1m, label='KBNC', marker='.', color='#9661BC')

ax.set(xlabel='Sampling', ylabel='kappa', ylim=[0.63, 0.72])

fig.legend()

plt.savefig('./readme.assets/Global_1m_kappa.png')
plt.show()

labels = ['KBNC', 'CABNC', 'CHILDES', 'BNC']
data = pd.concat([kbnc_kappa_df[0:1].transpose(), cabnc_kappa_df[0:1].transpose(), childes_kappa_df[0:1].transpose(), bnc_kappa_df[0:1].transpose()], axis=1, ignore_index=True)
data.columns = labels
plt.figure(figsize=(12,4))
plt.boxplot([data['KBNC'], data['CABNC'], data['CHILDES'], data['BNC']], labels=labels)
plt.savefig('./readme.assets/Global_1m_kappa_box.png')
plt.show()
```

```python
# balanced accuracy
kbnc_balancedacc_1m = np.array(kbnc_balanced_acc_df[0:1]).transpose()
cabnc_balancedacc_1m = np.array(cabnc_balanced_acc_df[0:1]).transpose()
childes_balancedacc_1m = np.array(childes_balanced_acc_df[0:1]).transpose()
bnc_balancedacc_1m = np.array(bnc_balanced_acc_df[0:1]).transpose()

fig, ax = plt.subplots()
fig.set_facecolor('white')
fig.set_figwidth(12)
x = np.array([x+1 for x in range(50)])
ax.plot(x, bnc_balancedacc_1m, label='BNC', marker='.', color='#5B8FF9')
ax.plot(x, childes_balancedacc_1m, label='CHILDS', marker='.', color='#61DDAA')
ax.plot(x, cabnc_balancedacc_1m, label='CABNC', marker='.', color='#F6BD16')
ax.plot(x, kbnc_balancedacc_1m, label='KBNC', marker='.', color='#9661BC')

ax.set(xlabel='Sampling', ylabel='balanced accuracy', ylim=[0.82, 0.87])

fig.legend()

plt.savefig('./readme.assets/Global_1m_acc.png')
plt.show()

labels = ['KBNC', 'CABNC', 'CHILDES', 'BNC']
data = pd.concat([kbnc_balanced_acc_df[0:1].transpose(), cabnc_balanced_acc_df[0:1].transpose(), childes_balanced_acc_df[0:1].transpose(), bnc_balanced_acc_df[0:1].transpose()], axis=1, ignore_index=True)
data.columns = labels
plt.figure(figsize=(12,4))
plt.boxplot([data['KBNC'], data['CABNC'], data['CHILDES'], data['BNC']], labels=labels)
plt.savefig('./readme.assets/Global_1m_acc_box.png')
plt.show()
```
